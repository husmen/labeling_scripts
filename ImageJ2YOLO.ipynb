```python
from os import walk, getcwd
from shutil import copyfile
from PIL import Image
import random
import pandas as pd
```

```python
classes = [
    "park",
    "girilmez",
    "ileri_saga",
    "calisma",
    "sola_donus",
    "yan_yol",
]

"""names = [
    "NO_PASS",
    "STOP",
    "DONT_ENTER",
    "ROADWORKS",
    "TRAFFIC_LIGHTS",
    "CROSSING",
    "RIGHT",
    "LEFT",
    "STRAIGHT",
    "STRAIGHT_OR_RIGHT",
    "STRAIGHT_OR_LEFT",
    "KEEP_RIGHT",
    "KEEP_LEFT",
]"""

names = [
    "PARK",
    "DONT_ENTER",
    "STRAIGHT_OR_RIGHT",
    "ROADWORKS",
    "RIGHT",
    "KEEP_LEFT",
]

images_dir = "014/"
labels_dir = "014_labels/"
output_dir = "traffic_signs/"
```

```python
# write class names
with open(output_dir + "class.names", "w") as f:
    for name in names:
        f.write(name + "\n")

# write data file        
with open(output_dir + "/traffic_signs.data", "w") as f:
    f.write("classes = " + str(len(classes)) + "\n")
    f.write("train = /home/lab/workspace/darknet/data/traffic_signs/train.txt" + "\n")
    f.write("valid = /home/lab/workspace/darknet/data/traffic_signs/test.txt" + "\n")
    f.write("names = /home/lab/workspace/darknet/data/traffic_signs/class.names" + "\n")
    f.write("backup = /home/lab/workspace/darknet/backup/" + "\n")
```

```python
df = pd.DataFrame(columns=['Path', 'ClassId', 'Roi.X1', 'Roi.Y1', 'Roi.X2', 'Roi.Y2'])
for root, dirs, files in walk(labels_dir):
    for file in files:
        if file.endswith("txt"):
            with open(labels_dir + file, "r") as f:
                lines = f.readlines()
                nmbr = file.split(".")[0]
                #img_path = images_dir + nmbr + ".jpg"
                #copyfile(img_path, output_dir + img_path.split("/")[-1])
                for line in lines:
                    tmp = line.split(" ")
                    df = df.append({'Path': nmbr, 'ClassId': tmp[0], 'Roi.X1': int(tmp[4]), 'Roi.Y1': int(tmp[5]), 'Roi.X2': int(tmp[6]), 'Roi.Y2': int(tmp[7])}, ignore_index=True)

print(len(df))
print(df[0:5])
```

```python
# preprocess training dataset
for i, index in enumerate(df.index):
    c = df.ClassId[index]
    if c == "ileri_sag" or c == "ieri_saga" or c == "ileri-saga":
        c = "ileri_saga"
    c = classes.index(c)
    n = df.Path[index]
    w = 1280
    h = 720
    x1 = df["Roi.X1"][index]
    y1 = df["Roi.Y1"][index]
    x2 = df["Roi.X2"][index]
    y2 = df["Roi.Y2"][index]
    
    ox = ((x1 + x2)/2)/w
    oy = ((y1 + y2)/2)/h
    ow = abs(x1 - x2)/w
    oh = abs(y1 - y2)/h
    
    string = str(c) + " " + str(ox) + " " + str(oy) + " " + str(ow) + " " + str(oh)
    
    print(n)
    print(string)
    with open(output_dir + n + ".txt", "a") as f:
        f.write(string + "\n")
        
random.seed(123)
file_history = set()
with open(output_dir + "train.txt", "w") as f_train, open(output_dir + "test.txt", "w") as f_test:
    for i, index in enumerate(df.index):
        n = df.Path[index]
        if n not in file_history:
            file_history.add(n)
            #copyfile(images_dir + n + ".jpg", output_dir + n + ".jpg")
            n = "data/traffic_signs/" + n + ".jpg"
            rnd = random.randint(0,5)
            if rnd == 0:
                f_test.write(n + "\n")
            else:
                f_train.write(n + "\n")
```
